```python
#Q2

Categories_list = df['category'].head(995).to_list()
Subscriber_Values_List = df['subscribers'].head(995).to_list()

# Convert each category to a string and then to uppercase
for i in range(len(Categories_list)):
    Categories_list[i] = str(Categories_list[i])
    Categories_list[i] = Categories_list[i].upper()

for i in range(len(Categories_list)-1,-1,-1):
    if Categories_list[i] == 'NAN' or math.isnan(Subscriber_Values_List[i]):
        Subscriber_Values_List.pop(i)
        Categories_list.pop(i)

# Count the frequency of each category
frequency_category = Counter(Categories_list)


# Initialize a dictionary with categories as keys and 0 as values
avgsub_dict = dict.fromkeys(Categories_list, 0)

# Sum the subscribers for each category
for i in range(len(Categories_list)):
    avgsub_dict[Categories_list[i]] += Subscriber_Values_List[i]

# Calculate the average number of subscribers for each category
for i in avgsub_dict:
    avgsub_dict[i] /= frequency_category[i]

# Sort the dictionary by the average number of subscribers in descending order
sorted_dict = dict(sorted(avgsub_dict.items(), key=lambda item: item[1], reverse=True))

# Extract sorted categories and their average subscriber counts
categories = list(sorted_dict.keys())
avg_subscribers = list(sorted_dict.values())

# Plot the data
plt.figure(figsize=(10, 6))
plt.barh(categories, avg_subscribers, color='skyblue')  # Create a horizontal bar chart
plt.xlabel('Average Number of Subscribers')  # Label the x-axis
plt.ylabel('Category')  # Label the y-axis
plt.title('Average Number of Subscribers per Category')  # Add a title to the plot
plt.gca().invert_yaxis()  # Invert the y-axis to display the highest average at the top
plt.show()  # Display the plot
```

```python
#Q5

Channel_Type_List = df['channel_type'].head(995).to_list()

# Convert each channel type to a string and then to uppercase
for i in range(len(Channel_Type_List)):
    Channel_Type_List[i] = str(Channel_Type_List[i])
    Channel_Type_List[i] = Channel_Type_List[i].upper()

# Convert the first 995 entries of the 'category' column to a list
Categories_list = df['category'].head(995).to_list()

# Convert each category to a string and then to uppercase
for i in range(len(Categories_list)):
    Categories_list[i] = str(Categories_list[i])
    Categories_list[i] = Categories_list[i].upper()

# Remove 'NAN' entries from both lists
for i in range(len(Categories_list) - 1, -1, -1):  # Iterate backwards to avoid index issues
    if Categories_list[i] == 'NAN' or Channel_Type_List[i] == 'NAN':
        Categories_list.pop(i)
        Channel_Type_List.pop(i)

# Initialize a dictionary with categories as keys and empty lists as values
Category_Channel_Type_Dict = {category: [] for category in Categories_list}

# Populate the dictionary with channel types for each category
for i in range(len(Categories_list)):
    category = Categories_list[i]
    channel_type = Channel_Type_List[i]
    Category_Channel_Type_Dict[category].append(channel_type)

# Count the frequency of each channel type for each category
for category in Category_Channel_Type_Dict:
    Category_Channel_Type_Dict[category] = Counter(Category_Channel_Type_Dict[category])

# Create a figure and axis for the plot
fig, ax = plt.subplots(figsize=(14, 8))

# Get a colormap with enough distinct colors
colors = cm.get_cmap('tab20', len(Category_Channel_Type_Dict))

# Plot the data for each category
for i, (category, counter) in enumerate(Category_Channel_Type_Dict.items()):
    labels = list(counter.keys())
    counts = list(counter.values())

    # Assign a distinct color to each bar
    bar_colors = [colors(i / len(Category_Channel_Type_Dict)) for _ in labels]

    ax.bar(labels, counts, color=bar_colors, label=category)

# Label the axes and add a title to the plot
ax.set_xlabel('Channel Type')
ax.set_ylabel('Count')
ax.set_title('Distribution of Channel Types per Category')

# Add a legend with category labels
ax.legend(title='Categories', bbox_to_anchor=(1, 1))

# Add gridlines to the y-axis
ax.grid(axis='y', linestyle='--', color='gray')

# Set y-axis ticks to be spaced every 5 units
ax.yaxis.set_major_locator(MultipleLocator(5))

# Rotate x-axis labels for better readability
plt.xticks(rotation=90)

# Adjust the layout to fit everything
plt.tight_layout()

# Display the plot
plt.show()
```

```python
#Q7

# Convert the first 995 entries of the 'highest_monthly_earnings', 'lowest_monthly_earnings', and 'category' columns to lists
Highest_Month_Earnings_list = df['highest_monthly_earnings'].head(995).to_list()
Lowest_Month_Earnings_list = df['lowest_monthly_earnings'].head(995).to_list()
Categories_list = df['category'].head(995).to_list()

# Convert each category to a string and then to uppercase
for i in range(len(Categories_list)):
    Categories_list[i] = str(Categories_list[i])
    Categories_list[i] = Categories_list[i].upper()

# Remove 'NAN' entries and entries where the highest or lowest earnings are NaN
for i in range(len(Categories_list) - 1, -1, -1):  # Iterate backwards to avoid index issues
    if Categories_list[i] == 'NAN' or math.isnan(Highest_Month_Earnings_list[i]) or math.isnan(Lowest_Month_Earnings_list[i]):
        Categories_list.pop(i)
        Highest_Month_Earnings_list.pop(i)
        Lowest_Month_Earnings_list.pop(i)

# Create dictionaries mapping categories to their respective highest and lowest monthly earnings
Highest_monthly_dict = dict(zip(Categories_list, Highest_Month_Earnings_list))
Lowest_Monthly_dict = dict(zip(Categories_list, Lowest_Month_Earnings_list))

# Sum the highest monthly earnings for each category
for i in range(len(Categories_list)):
    if not pd.isna(Highest_Month_Earnings_list[i]):  # Check if the highest earnings value is not NaN
        Highest_monthly_dict[Categories_list[i]] += Highest_Month_Earnings_list[i]

# Calculate the average highest monthly earnings for each category
for i in Highest_monthly_dict:
    Highest_monthly_dict[i] /= frequency_category[i]

# Sum the lowest monthly earnings for each category
for i in range(len(Categories_list)):
    if not pd.isna(Lowest_Month_Earnings_list[i]):  # Check if the lowest earnings value is not NaN
        Lowest_Monthly_dict[Categories_list[i]] += Lowest_Month_Earnings_list[i]

# Calculate the average lowest monthly earnings for each category
for i in Lowest_Monthly_dict:
    Lowest_Monthly_dict[i] /= frequency_category[i]

# Sort the dictionaries by the averaged earnings in descending order
sorted_dict_1 = dict(sorted(Highest_monthly_dict.items(), key=lambda item: item[1], reverse=True))
sorted_dict_2 = dict(sorted(Lowest_Monthly_dict.items(), key=lambda item: item[1], reverse=True))

# Extract sorted categories and their averaged highest and lowest earnings
categories = list(sorted_dict_1.keys())
highest_earnings = list(sorted_dict_1.values())
lowest_earnings = list(sorted_dict_2.values())

# Create a figure for the plot
plt.subplots(figsize=(15, 8))

# Set the width of the bars
bar_width = 0.35
index = range(len(categories))

# Plot the averaged highest monthly earnings
plt.bar(index, highest_earnings, bar_width, color='g', label='Averaged Highest Monthly Earnings')

# Plot the averaged lowest monthly earnings
plt.bar([i + bar_width for i in index], lowest_earnings, bar_width, color='r', label='Averaged Lowest Monthly Earnings')

# Label the axes and add a title to the plot
plt.xlabel('Categories')
plt.ylabel('Earnings')
plt.title('Averaged Highest and Lowest Monthly Earnings by Category')

# Set the x-ticks to the center of the grouped bars and rotate labels for better readability
plt.xticks([i + bar_width / 2 for i in index], categories, rotation=90)

# Add a legend to the plot
plt.legend()

# Adjust the layout to fit everything
plt.tight_layout()

# Display the plot
plt.show()
```

```python
#Q9

# Extract the first 995 entries of the 'Youtuber' column to a list and convert each entry to a string
Youtuber_list = df['Youtuber'].head(995).to_list()
for i in range(len(Youtuber_list)):
    Youtuber_list[i] = str(Youtuber_list[i])

# Extract the first 995 entries of the 'highest_yearly_earnings' column to a list
highest_yearly_earnings_list = df['highest_yearly_earnings'].head(995).to_list()

# Remove entries where the highest yearly earnings are zero, NaN, or the YouTuber name is 'nan'
for i in range(len(highest_yearly_earnings_list) - 1, -1, -1):  # Iterate backwards to avoid index issues
    if highest_yearly_earnings_list[i] == 0 or Youtuber_list[i] == 'nan' or math.isnan(highest_yearly_earnings_list[i]):
        Youtuber_list.pop(i)
        highest_yearly_earnings_list.pop(i)

# Extract the first 995 entries of the 'lowest_yearly_earnings' column to a list
lowest_yearly_earnings_list = df['lowest_yearly_earnings'].head(995).to_list()

# Re-extract the first 995 entries of the 'Youtuber' column to a list and convert each entry to a string
Youtuber_list = df['Youtuber'].head(995).to_list()
for i in range(len(Youtuber_list)):
    Youtuber_list[i] = str(Youtuber_list[i])

# Remove entries where the lowest yearly earnings are zero, NaN, or the YouTuber name is 'nan'
for i in range(len(lowest_yearly_earnings_list) - 1, -1, -1):  # Iterate backwards to avoid index issues
    if lowest_yearly_earnings_list[i] == 0 or Youtuber_list[i] == 'nan' or math.isnan(lowest_yearly_earnings_list[i]):
        Youtuber_list.pop(i)
        lowest_yearly_earnings_list.pop(i)

# Create dictionaries mapping YouTubers to their respective highest and lowest yearly earnings
Highest_earners_dic = dict(zip(Youtuber_list, highest_yearly_earnings_list))
Lowest_earners_dic = dict(zip(Youtuber_list, lowest_yearly_earnings_list))

# Sort the dictionaries by the earnings in descending order
sorted_dict_1 = dict(sorted(Highest_earners_dic.items(), key=lambda item: item[1], reverse=True))
sorted_dict_2 = dict(sorted(Lowest_earners_dic.items(), key=lambda item: item[1], reverse=True))

# Define a function to plot the top 5 and bottom 5 earners from a dictionary
def plot_first_last_five(dictionary, title):
    keys = list(dictionary.keys())
    values = list(dictionary.values())

    # Get the first and last 5 entries
    first_five_keys = keys[:5]
    first_five_values = values[:5]
    last_five_keys = keys[-5:]
    last_five_values = values[-5:]

    # Create subplots for the top 5 and bottom 5 earners
    fig, axs = plt.subplots(2, 1, figsize=(10, 8))

    # Plot the top 5 earners
    axs[0].bar(first_five_keys, first_five_values, color='green')
    axs[0].set_title(f'Highest {title} yearly earnings')
    axs[0].set_ylabel('Values')

    # Plot the bottom 5 earners
    axs[1].bar(last_five_keys, last_five_values, color='red')
    axs[1].set_title(f'Lowest {title} yearly earnings (non-zero)')
    axs[1].set_ylabel('Values')

    # Adjust the layout and show the plot
    plt.tight_layout()
    plt.xticks(rotation=60)
    plt.show()

# Plot the top and bottom 5 highest yearly earnings
plot_first_last_five(sorted_dict_1, 'maximum')

# Plot the top and bottom 5 lowest yearly earnings
plot_first_last_five(sorted_dict_2, 'minimum')
```

```python
#Q10
# Mapping of month abbreviations to their numerical values
month_mapping = {
    "Jan": 1, "Feb": 2, "Mar": 3, "Apr": 4, "May": 5, "Jun": 6,
    "Jul": 7, "Aug": 8, "Sep": 9, "Oct": 10, "Nov": 11, "Dec": 12
}

# Extract the first 995 entries of the 'created_year', 'created_month', and 'created_date' columns to lists
channel_created_year_list = df['created_year'].head(995).to_list()
channel_created_month_list = df['created_month'].head(995).to_list()
channel_created_day_list = df['created_date'].head(995).to_list()

# Remove entries where the year, month, or day is NaN or the month is 'nan'
for i in range(len(Youtuber_list) - 1, -1, -1):  # Iterate backwards to avoid index issues
    if pd.isna(channel_created_year_list[i]) or channel_created_month_list[i] == 'nan' or pd.isna(channel_created_month_list[i]):
        channel_created_year_list.pop(i)
        channel_created_month_list.pop(i)
        channel_created_day_list.pop(i)

# Convert month abbreviations to their numerical values using the month_mapping dictionary
channel_created_month_list = [month_mapping[month] for month in channel_created_month_list]

# Create a list of tuples (year, month, day)
dates_list = list(zip(channel_created_year_list, channel_created_month_list, channel_created_day_list))

# Convert the tuples to integers
for i in range(len(dates_list)):
    dates_list[i] = (int(dates_list[i][0]), int(dates_list[i][1]), int(dates_list[i][2]))

# Create a list of Timestamps from the dates
dates = [pd.Timestamp(year=y, month=m, day=d) for y, m, d in dates_list]

# Create a scatter plot of the dates vs. their ranks
plt.figure(figsize=(10, 6))
plt.scatter(dates, range(len(dates)), alpha=0.6)
plt.xlabel('Date')
plt.ylabel('Rank')
plt.title('Channel Creation Dates vs Rank')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()

# Display the plot
plt.show()
```

```python
#Q12

# Load the data and prepare the lists
country_list = df['Country'].head(995).to_list()
unemployment_rate_list = df['Unemployment rate'].head(995).to_list()

# Convert country names to uppercase
for i in range(len(country_list)):
    country_list[i] = str(country_list[i])
    country_list[i] = country_list[i].upper()



# Remove entries with NaN unemployment rates and corresponding countries
for i in range(len(country_list)-1, -1, -1):
    if pd.isna(unemployment_rate_list[i]) or country_list[i] == 'NAN':
        unemployment_rate_list.pop(i)
        country_list.pop(i)
# Calculate the frequency of each country
frequency_country = Counter(country_list)

# Create a dictionary for unemployment rates and calculate the total unemployment rate for each country
unemployment_dict = dict.fromkeys(country_list, 0)
for i in range(len(country_list)):
    unemployment_dict[country_list[i]] += unemployment_rate_list[i]

# Calculate the average unemployment rate for each country
for i in unemployment_dict:
    unemployment_dict[i] /= frequency_country[i]

# Sort the dictionary by average unemployment rate in descending order
sorted_unemployment_dict = dict(sorted(unemployment_dict.items(), key=lambda item: item[1], reverse=True))

# Get the top 10 countries with the highest average unemployment rate
top_unemployment_countries = list(sorted_unemployment_dict.items())[:10]

top_unemployment_country_names = [country[0] for country in top_unemployment_countries]
top_unemployment_country_rates = [country[1] for country in top_unemployment_countries]

# Plot the top 10 countries by average unemployment rate
plt.figure(figsize=(12, 8))
plt.bar(top_unemployment_country_names, top_unemployment_country_rates, color='blue')
plt.xlabel('Country')
plt.ylabel('Average Unemployment Rate')
plt.title('Top 10 Countries by Average Unemployment Rate')
plt.xticks(rotation=45)
plt.show()

# Get the top 10 countries with the highest number of YouTube channels
top_channel_countries = frequency_country.most_common(10)

top_channel_country_names = [country[0] for country in top_channel_countries]
top_channel_country_counts = [country[1] for country in top_channel_countries]

# Sort the unemployment rates of these top channel countries in the same order
top_channel_country_unemployment_rates = [unemployment_dict[country] for country in top_channel_country_names]

# Plot the top 10 countries by number of YouTube channels
plt.figure(figsize=(12, 8))
plt.bar(top_channel_country_names, top_channel_country_counts, color='green')
plt.xlabel('Country')
plt.ylabel('Number of YouTube Channels')
plt.title('Top 10 Countries by Number of YouTube Channels')
plt.xticks(rotation=45)
plt.show()

# Plot the unemployment rates for the top 10 countries with the highest number of YouTube channels
plt.figure(figsize=(12, 8))
plt.bar(top_channel_country_names, top_channel_country_unemployment_rates, color='red')
plt.xlabel('Country')
plt.ylabel('Average Unemployment Rate')
plt.title('Average Unemployment Rate for Top 10 Countries by Number of YouTube Channels')
plt.xticks(rotation=45)
plt.show()
```

```python
# Mapping for month abbreviation to numerical month
month_mapping = {
    'JAN': 1, 'FEB': 2, 'MAR': 3, 'APR': 4,
    'MAY': 5, 'JUN': 6, 'JUL': 7, 'AUG': 8,
    'SEP': 9, 'OCT': 10, 'NOV': 11, 'DEC': 12
}

# Extracting data from the DataFrame
create_month_list = df['created_month'].head(995).to_list()
create_year_list = df['created_year'].head(995).to_list()
subscribes_list = df['subscribers'].head(995).to_list()

# Convert month abbreviations to uppercase
for i in range(len(create_month_list)):
    create_month_list[i] = str(create_month_list[i]).upper()

# Removing invalid entries
for i in range(len(create_month_list) - 1, -1, -1):
    if (create_month_list[i] == 'NAN' or
        create_month_list[i] not in month_mapping or
        math.isnan(create_year_list[i]) or
        math.isnan(subscribes_list[i])):
        create_month_list.pop(i)
        create_year_list.pop(i)
        subscribes_list.pop(i)

# Map month abbreviations to numerical month
create_month_list = [month_mapping[month] for month in create_month_list]

# Calculate total number of months from creation date to 2023
Total_month_list = [((2023 - create_year_list[i])*12) + create_month_list[i] for i in range(len(create_month_list))]

# Calculate average subscribers per month
avg_sub_per_month = [subscribes_list[i] / Total_month_list[i] for i in range(len(Total_month_list))]

# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(avg_sub_per_month, marker='o', linestyle='-', color='b')
plt.xlabel('Rank of Channels')
plt.ylabel('Average Subscribers per Month from creation')
plt.title('Average Subscribers per Month by Rank of Channels')
plt.grid(True)
plt.show()
```
